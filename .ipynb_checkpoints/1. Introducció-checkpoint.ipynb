## Per què hem escollit *jupyter notebook* ?

*jupyter notebook* és un entorn de treball Python (encara que pot treballar amb molts més llenguatges) amb una gran potència de càlcul i de visualització de dades, i que permet la interactivitat amb el lector. 

Justament la promoció que farà 1r de Batxillerat a l'institut *Príncep de Viana* és la primera promoció que va treballar amb micropython a 3r d'ESO. Per tant estaran familizaritzats amb el llenguatge de programació.

El projecte *jupiter* és una organització sense ànim de lucre creada per desenvolupar codi lliure, estàndars oberts i serveis per a computació interactiva en dotzenes de llenguatges de programació. Per tant podem instal·lar lliurement *jupyter notebook* al nostre ordinador, i així crear els documents. 

Per altra banda aquests llibres interactius es podem consultar lliurament si els podem a *github* i els registrem a https://mybinder.org/. Com *binder* treballa llegint el nostre *github*, qualsevol modificació que els usuaris puguin fer al document és una modificació local, que no afecta al document original. El projecte *binder* forma part de *jupyter* i ofereix gratuïtament l'accés simultani de fins a 100 usuaris al nostre document al seu servidor. 

A més a més amb *jupyter notebook* tenim:
- La possibilitat d'escriure text amb la codificació Markdown, com el que esteu ara llegint
- La possibilitat d'incloure imatges, fins-i-tot externes. Es pot fer amb Markdown, directament. 
- També podem incloure vídeos
- Podem codificar fòrmules amb Latex
- Podem mostrar dades amb un format agradable amb HTML
- Tenim al nostre abast funcions matemàtiques potents amb la llibreria *numpy*
- Podem generar gràfics amb la llibreria *matplotlib*
- Tenim accés a la llibreria  de tractament algebraic *sympy*

### Exemple fòrmula amb Latex

```python
%%latex
\begin{align}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\
\nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0
\end{align}
```

### Exemple gràfica

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x**2)
plt.plot(x, y)
plt.title("A little chirp")
fig = plt.gcf() 
```

### Exemple algebraic

```python
from IPython.display import display
from sympy.interactive import printing
printing.init_printing(use_latex='mathjax')
from sympy import *
x, y, z = symbols("x y z")

eq = ((x+y)**2 * (x+1))
eq
```

```python
expand(eq)
```

```python
eq = Eq(x**3 - 8*x**2 + 5*x + 14, 0)
eq
```

```python
solve(eq, x)
```
